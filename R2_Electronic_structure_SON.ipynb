```python
def get_vbm(calc):
    occ = calc.occupations
    eig = calc.eigenvalues
    nkpts, nbands = np.shape(occ)
    vbm = -100
    for k in range(nkpts):
        for nb in range(nbands):
            if occ[k][nb] == 2 and eig[k][nb]>vbm:
                vbm = eig[k][nb]
    return vbm
                
```

```python
def get_one_pdos_dos(path):
    calc_file = os.path.join(path, 'OUTCAR')
    dos_file = os.path.join(path, 'DOSCAR')
    calc = Extract(calc_file)
    vol = float(calc.volume)
    dos = VaspDos(dos_file)
    energy = dos.energy # -  float(calc.vbm)
    #print calc.vbm
    Ngrid = len(dos.dos)
    Natoms = len(calc.structure)
    pdos = np.zeros([Natoms,Ngrid])
    for i in range(Natoms):
        for id in range(9):
            pdos[i] += dos.site_dos(i, id)
    #print np.shape(pdos)
    return energy, pdos, 1e3*dos.dos/vol, float(get_vbm(calc)), float(calc.cbm)
```

```python
def calc_ipr(pdos):
    Natoms, Nel = np.shape(pdos)
    c2 = pdos * pdos
    c4 = c2 * c2
    ipr = Natoms * np.sum(c4, axis=0)/np.sum(c2, axis=0)**2
    return ipr
#ipr = np.nan_to_num(ipr)
```

```python
def clean(ipr):
    N = len(ipr)
    i_tmp = np.zeros(N)
    for i in range(N):
        if ipr[i] == np.nan:
            i_tmp[i] = (ipr[i+1]+ipr[i-1])/2
        else:
            i_tmp[i] = ipr[i]
    return i_tmp
```
