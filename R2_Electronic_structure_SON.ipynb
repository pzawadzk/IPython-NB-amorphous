```python
def get_vbm(calc):
    """ Calculates vbm """
    occ = calc.occupations
    eig = calc.eigenvalues
    nkpts, nbands = np.shape(occ)
    vbm = -100
    for k in range(nkpts):
        for nb in range(nbands):
            if occ[k][nb] == 2 and eig[k][nb]>vbm:
                vbm = eig[k][nb]
    return vbm    
```

```python
def check_for_AA_bonds(structure):
    cut_off = {'O':2.0, 'N':2.0, 'S':2.5}
    AA = []
    for index_A, atom in enumerate(structure):
        if atom.type in cut_off.keys():
            pos_A = structure[index_A].pos
            neighs = [n for n in neighbors(structure, 10, pos_A)]
            for n in neighs:
                if n[0].type in cut_off.keys():
                  if n[2]*structure.scale < cut_off[atom.type]: AA.append(index_A)
    return AA
```

```python
def Extract_DOS(path):
    """ Extracts denisty of states, normalization factor, structure and valence band maximum """
    calc_file = os.path.join(path, 'OUTCAR')
    dos_file = os.path.join(path, 'DOSCAR')
    calc = Extract(calc_file)
    vol = float(calc.volume)
    
    AA_bonds = check_for_AA_bonds(calc.structure)
    
    Natoms = len(calc.structure)
    DOS = VaspDos(dos_file)
    mult = 1e3/vol
    
    dos = DOS.dos * mult
    Ngrid = len(dos)
    pdos = np.zeros([Natoms, 9, Ngrid])
    
    for i in range(Natoms):
        for id in range(9):
            pdos[i][id] += DOS.site_dos(i, id)
    pdos *= mult
    return DOS.energy, dos, pdos, calc.structure, float(calc.vbm), float(calc.cbm), AA_bonds
```

```python
def get_dos_data(path, key, avg=True):

    paths = glob.glob(path%key)
    
    DOS, PDOS, PDOS_C, PDOS_A, PDOS_AA, VBM, CBM  = [], [],[], [], [], [], []
    E = None
    for path in paths:
        print path
        try:
            E, dos, pdos, structure, vbm, cbm, AA_bonds = Extract_DOS(path)
            A_atoms, C_atoms = [], []
            for i in range(len(structure)):
                if structure[i].type in ['O', 'N', 'S']:
                    if i not in AA_bonds: A_atoms.append(i)
                else: C_atoms.append(i)
            
            DOS.append(dos)
            PDOS.append(pdos)
            PDOS_C.append(np.mean(pdos[C_atoms, :, :], axis=0))
            PDOS_A.append(np.mean(pdos[A_atoms, :, :], axis=0))
            PDOS_AA.append(np.mean(pdos[AA_bonds, :, :], axis=0))

            VBM.append(vbm)
            CBM.append(cbm)
            print 'DONE in', key
        except:
            pass
        
    if avg:
        
        DOS = np.mean(DOS, axis = 0)
        PDOS = np.mean(np.mean(PDOS, axis = 0), axis=1)
        
        PDOS_A = np.mean(PDOS_A, axis = 0)
        PDOS_AA = np.mean(PDOS_AA, axis = 0)
        PDOS_C = np.mean(PDOS_C, axis = 0)
        
        VBM = np.mean(VBM, axis = 0)
        CBM = np.mean(CBM, axis = 0)
    return E, DOS, PDOS, PDOS_A, PDOS_AA, PDOS_C, VBM, CBM
```

```python
def calc_ipr(pdos):
    Natoms, Nel = np.shape(pdos)
    c2 = pdos * pdos
    c4 = c2 * c2
    ipr = Natoms * np.sum(c4, axis=0)/np.sum(c2, axis=0)**2
    return ipr
#ipr = np.nan_to_num(ipr)
```

```python
def clean(ipr):
    N = len(ipr)
    i_tmp = np.zeros(N)
    for i in range(N):
        if ipr[i] == np.nan:
            i_tmp[i] = (ipr[i+1]+ipr[i-1])/2
        else:
            i_tmp[i] = ipr[i]
    return i_tmp
```
