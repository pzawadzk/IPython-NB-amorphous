#Analysis of Antimoney oxynitrides

```python
from dist_func import calc_pdfs,  int_cn_from_pdfs, calc_adfs, get_cn
import pickle
import sys
import os
import numpy as np
from pylada.crystal import read, supercell
from pylada.vasp import  Extract
%pylab inline
from pylada.crystal import neighbors
```

```python
def get_cn_ON(structure,  cut_off = 2.8):
    
    scale = float(structure.scale)
    cn = np.zeros(15)
    for index_A in range(len(structure)):
      A = structure[index_A].type
      if A !='O' and  A !='N':
          pos_A = structure[index_A].pos
          neighs = [n for n in neighbors(structure, 14, pos_A)]
          count = 0
          for i, nB in enumerate(neighs):
            B = nB[0].type
            if B == 'O' or B == 'N':
                d2_AB = (scale * nB[2])
                d2_AB_prev = d2_AB
                if d2_AB > cut_off:
                  break
                else:
                  count += 1
          #print count
          cn[count] += 1
    return cn
```

```python
data_XO = np.zeros(121)
data_XN = np.zeros(121)
data_OO = np.zeros(121)
data_ON = np.zeros(121)
data_NN = np.zeros(121)
data_cn = np.zeros(15)
CN_all = []
seeds = [2]
for seed in seeds:
    file = '/scratch/pzawadzk/amor/SbON/bulk/126_seed_%d_par_A_4_B_13_T0_sc1.00_dist_con/basin_hoping/OUTCAR'%seed
    if os.path.isfile(file):
        calc = Extract(file)
        structure = calc.structure
        CN = get_cn_ON(structure, cut_off=2.8)
        P_temp, r = calc_pdfs(structure, pdf_cut = 6., pdf_sigma=0.2)
        CN_all.append(get_cn(structure, cut_off=1.9))
        data_XO += P_temp['SbO']/len(seeds)
        data_XN += P_temp['SbN']/len(seeds)
        data_OO += P_temp['OO']/len(seeds)
        data_NN += P_temp['NN']/len(seeds)
        data_ON += P_temp['ON']/len(seeds)
        data_cn += np.array(CN)/len(seeds)
```

```python
figure()
ax = gca()
plot(r, data_XO, '-r', label = 'SbO')
plot(r, data_XN, '-b', label = 'SbN')
ax.set_xlabel(r'distance [$\AA$]')
ax.set_ylabel(r'pair distribution function')
ax.axis(xmin=0, xmax=5.5, ymin=0, ymax=3.5)
legend()

figure()
ax = gca()
plot(r, data_OO, '-r', label = 'OO')
plot(r, data_NN, '-b', label = 'NN')
plot(r, data_ON, '-k', label = 'ON')
ax.set_xlabel(r'distance [$\AA$]')
ax.set_ylabel(r'pair distribution function')
ax.axis(xmin=0, xmax=5.5, ymin=0, ymax=2.5)
legend()


figure()
ax = gca()
plot(data_cn/np.sum(data_cn), '-k', label= 'Sb-{O,N}$_k$')
ax.set_xlabel(r'coordination number, $k$')
ax.set_ylabel(r'fraction of atoms')
ax.axis(xmin=0, xmax=10, ymin=0, ymax=0.8)
legend()
print len(r)
```
