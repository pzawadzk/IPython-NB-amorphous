## Pair Distribution Function

```python
from dist_func import calc_pdfs,  int_cn_from_pdfs, calc_adfs, get_cn
import pickle
import sys
import os
import numpy as np
from pylada.crystal import read, supercell
from pylada.vasp import  Extract
%pylab inline
```

```python

materials = {
             'In2O3':[-2,-1,1,40,0],
             'ZnO':[-2, -1,1,30,0],
      'Si': [-2,-1,0.8, 32,0],
        }
```

```python
#Processed data
data_dir = 'data_p101_1'
if not os.path.exists(data_dir):
    os.makedirs(data_dir)   
data_file = os.path.join(data_dir, 'data_structure')
```

```python
cut ={'In2O3':2.9,
      'Si':2.9,
      'ZnO':2.7}
if os.path.exists(data_file):
    f = open(data_file, 'r')
    P_dict = pickle.load(f)
    f.close()
else:
  P_dict ={}
  A_dict ={}
  for name in materials.keys():
    for Q in materials[name]:
        count = 0
        P = {}
        A = {}
        CN = {}
        CN2 = {}
        if Q == 0:
            seeds = range(1,5)
        else:
            seeds = [1,]
        for seed in seeds:
            if name == 'In2O3':
                if Q == 0:
                    dir0 = 'rs_data/In2O3_ZnO'
                    dir = '160_seed_%d_par_in2o3_32_zno_0_T%d' % (seed, Q)
                    file = dir0 + '/' + dir + '/rls/OUTCAR'
                elif Q == -1:
                    dir0 = 'bulk/In2O3'
                    dir = 'In2O3'
                    file = dir0 + '/OUTCAR'    
                elif Q == -2:
                    dir0 = 'sa_data/In2O3'
                    dir = 'cryst_melt'
                    file = dir0 + '/' + dir + '/OUTCAR'
                else:
                    dir0 = 'sa_data/In2O3'
                    dir = '160_cryst_rho_c_15_%d_' % ( Q)
                    file = dir0 + '/' + dir + '/simulated_annealing/OUTCAR'
            elif name == 'ZnO':
                if Q == 0:
                    dir0 = 'rs_data/ZnO'
                    dir = '160_seed_%d_par_in2o3_0_zno_80_T%d' % (seed, Q)
                    file = dir0 + '/' + dir + '/rls/OUTCAR'
                elif Q == -2:
                    dir0 = 'sa_data/ZnO'
                    dir = 'cryst_melt'
                    file = dir0 + '/' + dir + '/OUTCAR'                
                elif Q == -1:
                    dir0 = 'bulk/ZnO'
                    file = dir0 + '/OUTCAR'    
                else:
                    dir0 = 'sa_data/ZnO'
                    dir = '160_seed_1_par_15_%d_' % ( Q)
                    file = dir0 + '/' + dir + '/simulated_annealing/OUTCAR'
            elif name == 'Si':
                if Q == 0:
                    dir0 = 'rs_data/Si'
                    dir = '160_seed_%d_par_rho_c' % (seed)
                    file = dir0 + '/' + dir + '/rls/OUTCAR'
                elif Q == -1:
                    dir0 = 'bulk/Si'
                    dir = 'cryst'
                    file = dir0 + '/' + dir + '/OUTCAR'
                elif Q == -2:
                    dir0 = 'sa_data/Si'
                    dir = 'melt'
                    file = dir0 + '/' + dir + '/OUTCAR'
                else:
                    dir0 = 'sa_data/Si'
                    dir = '160_seed_1_par_rho_c_15.0_%.1f_' % ( Q)
                    file = dir0 + '/' + dir + '/simulated_annealing/OUTCAR'

            file = os.path.join('/scratch/pzawadzk/amor_rmesa', file)
            print file
            if os.path.isfile(file):
                calc = Extract(file)
                structure = calc.structure
                CN2_temp = get_cn(structure, cut_off=cut[name])
                if len(structure)<50:
                    structure = supercell(structure, np.dot(structure.cell, np.eye(3)*3))
                count +=  1
                if Q == -1:
                    P_temp, r = calc_pdfs(structure, pdf_cut = 10., pdf_sigma=0.01)
                else:
                    P_temp, r = calc_pdfs(structure, pdf_cut = 10., pdf_sigma=0.2)
                CN_temp = int_cn_from_pdfs(structure, P_temp)
                A_temp = calc_adfs(structure)
                
                for type in P_temp.keys():
                    if type not in P.keys():
                        P[type]  = P_temp[type]
                        CN[type]  = CN_temp[type]
                        CN2[type]  = CN2_temp[type]
                    else:
                        P[type]  += P_temp[type]
                        CN[type]  += CN_temp[type]
                        CN2[type]  += CN2_temp[type]
                for type in A_temp.keys():
                    if type not in A.keys():
                        A[type]  = A_temp[type]
                    else:
                        A[type]  += A_temp[type]
     
        for type in A.keys():
            A[type]  = A[type]/count
            
        for type in P.keys():
            P[type]  = P[type]/count
            CN[type]  = CN[type]/count
            CN2[type]  = CN2[type]/count
            
        if name not in P_dict.keys():   
            P_dict[name]=[]
        P_dict[name].append([P, CN, r, A, CN2])
        
  f = open(data_file, 'w')
  pickle.dump(P_dict,f)
  f.close()
   
```
