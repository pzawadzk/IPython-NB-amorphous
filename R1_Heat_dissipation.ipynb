```python
#Processed data
data_dir = 'data_p101_1'
if not os.path.exists(data_dir):
    os.makedirs(data_dir)   
heat_disp_file = os.path.join(data_dir, 'heat_disp')

```

```python
def get_ms_atom_disp(structure0, structure):
    r2 = []

    for index in range(len(structure0)):
      atom = structure[index].type
      dp = structure[index].pos - structure0[index].pos
      dp  = impose_pbc(structure.cell, dp)
      r2.append(np.dot(dp,dp)*structure.scale**2)

    return np.array(r2)

def calc_msd(strs, N):
  [str_0] = read_xdatcar.get_structures(path = '/scratch/pzawadzk/amor_rmesa/heat_dis/In2O3/microcan/POSCAR')
  msd = np.zeros([N, 160])
  for i, str in enumerate(strs[:N]):
    r2 = get_ms_atom_disp(str_0, str)**0.5
    msd[i] = r2
  return msd

def runningMeanFast(x, N):
    return np.convolve(x, np.ones((N,))/N)[(N-1):]
    
```
