```python
def get_data(key, last = True):
    pdf = []
    cn = []
    if last:
        paths = glob.glob('/scratch/pzawadzk/amor_bulk_hf/%s/bulk/*_B_0_*'%key)
    else:
        paths = glob.glob('/scratch/pzawadzk/amor_bulk_hf/%s/bulk/*_A_0_*'%key)
    for path in paths:
        print path
        try:
            calc = Extract(os.path.join(path, 'basin_hoping/OUTCAR'))
            P_temp, r = calc_pdfs(calc.structure, pdf_cut = 10.1, pdf_sigma=0.1)
            print cut[key]
            pdf.append(P_temp)
            cn.append(get_cn(calc.structure, cut_off=cut[key]))
        except:
            print 'error'
            pass
    return pdf, cn, r
```
